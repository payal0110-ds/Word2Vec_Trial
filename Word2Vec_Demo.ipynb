```python
import gensim.downloader as api
wv = api.load('word2vec-google-news-300')
vec_king = wv['king']
print(vec_king)
```

```python
print(vec_king.shape)
print(len(vec_king))
```

```python
print(wv['mango'])
```

```python
wv.most_similar('mango')
```

```python
wv.most_similar('videogame')
```

```python
print(wv.similarity('man','boy'))
```

```python
print(wv.similarity('fruit','apple'))
```

```python
vec_res=wv['king']-wv['man']+wv['woman']
vec_res
```

```python
wv.most_similar([vec_res])
```
